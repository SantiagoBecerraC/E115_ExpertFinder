
`search_people` method:
- keywords (str, optional) ‚Äì Keywords to search on
- current_company (list, optional) ‚Äì A list of company URN IDs (str)
- past_companies (list, optional) ‚Äì A list of company URN IDs (str)
- regions (list, optional) ‚Äì A list of geo URN IDs (str)
- industries (list, optional) ‚Äì A list of industry URN IDs (str)
- schools (list, optional) ‚Äì A list of school URN IDs (str)
- profile_languages (list, optional) ‚Äì A list of 2-letter language codes (str)
- contact_interests (list, optional) ‚Äì A list containing one or both of ‚ÄúproBono‚Äù and ‚ÄúboardMember‚Äù
- service_categories (list, optional) ‚Äì A list of service category URN IDs (str)
- network_depth (str, optional) ‚Äì Deprecated, use network_depths. One of ‚ÄúF‚Äù, ‚ÄúS‚Äù and ‚ÄúO‚Äù (first, second and third+ respectively)
- network_depths (list, optional) ‚Äì A list containing one or many of ‚ÄúF‚Äù, ‚ÄúS‚Äù and ‚ÄúO‚Äù (first, second and third+
respectively)
- include_private_profiles (boolean, optional) ‚Äì Include private profiles in search results. If False, only public profiles are included. Defaults to False
- keyword_first_name (str, optional) ‚Äì First name
- keyword_last_name (str, optional) ‚Äì Last name
- keyword_title (str, optional) ‚Äì Job title
- keyword_company (str, optional) ‚Äì Company name
- keyword_school (str, optional) ‚Äì School name
- connection_of (str, optional) ‚Äì Connection of LinkedIn user, given by profile URN ID
- limit (int, optional) ‚Äì Maximum length of the returned list, defaults to -1 (no limit)

```python
# Search for software engineers at Google
people = api.search_people(
    keywords="machine learning computer vision",
    regions = ['103644278',],
    limit=100
)

people
```
